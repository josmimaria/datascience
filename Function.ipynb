```python
def test7(a):
  n=[]
  for i in a:
    if type(i)==list:
      for j in i:
        if type(j)==int or type(j)== float:
          n.append(j)
    else:
      if type(i)==int or type(i)==float:
        n.append(i)
  return n
```

```python
test7(l)
```

```python
def test11(*args):
  n=[]
  for i in args:
    if type(i)==list:
      n.append(i)
  return n
```

```python
# **kwargs is used in Python functions to accept a variable number of keyword arguments.
#The arguments are stored as a dictionary inside the function.
def test12(**kwargs):
  return kwargs
```

```python
test12(a="manu",b=32,c=[1,2,3,4,5],d=("data","ML"),e=True)
```

```python
def test13(**kwargs):
  for i in kwargs:
    if type(kwargs[i])== list:
      return {i: kwargs[i]}
```

```python
test13(a="manu",b=32,c=[1,2,3,4,5],d=("data","ML"),e=True)
```

```python
def test14(*args, **kwargs): # *args return as Tuple and **kwargs return as a dictionary
  return(args,kwargs)
```

```python
test14(1,2,3,4,5,a="test",b="data")
```
