# Generator Function

**yield** is used in Python generator functions to produce values one at a time and pause the function’s execution between each value, maintaining its state for the next iteration.

**Difference between return and yield**

return → stops function completely

yield → pauses and continues later

```python
def test_fib(n): # It is generator function, its return only generator object
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b
```

```python
test_fib(10)
```

```python
for i in test_fib(10):
  print(i)
```

```python
def gen_fun1(n):
  i=1;
  while n>=i:
    yield i
    i+=1
```

```python
for i in gen_fun1(5):
  print(i)
```

```python
def gen_fun2(n):
  #square_lst=[]
  for i in range(n):
    yield i*i
```

```python
square_lst=[]
for i in gen_fun2(5):
  square_lst.append(i)
```

```python
square_lst
```

```python
print(list(gen_fun2(5)))
```

# Lambda Function

A lambda function in Python is a small anonymous function (without a name) that can take any number of arguments but has only one expression. It is used for simple operations where defining a full function using def is not necessary.

We use lambda functions because they are short, quick to write, and often used with functions like map(), filter(), and sorted() for simple transformations or calculations.

```python
def test_l1(n,p):
  return n**p
```

```python
test_l1(3,2)
```

```python
test_l1(2,3)
```

```python
a = lambda n,p : n**p
```

```python
a(3,2)
```

```python
a(3,3)
```
