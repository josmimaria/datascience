# Stack

A stack is a linear data structure that follows:

LIFO → Last In, First Out

The last element added is removed first.

# Example:

Stack of plates 🍽️
Last plate placed on top is removed first.

# Operations
**push()**→ add element

**pop()** → remove top element

**peek()** → view top element

```python
stack = []

# Push
stack.append(10)
stack.append(20)
stack.append(30)

print(stack)

# peek
top_element = stack[-1]

print("Top element:", top_element)
# Pop
stack.pop()

print(stack)
```

# Queue

A queue is a linear data structure that follows:

FIFO → First In, First Out

The first element added is removed first.

# Example:

People standing in a line 🚶.

First person in line exits first

# Operations

**enqueue()** → add element

**dequeue()** → remove front element

```python
from collections import deque

queue = deque()


# Enqueue
queue.append(10)
queue.append(20)
queue.append(30)

print(queue)

# Dequeue
queue.popleft()

print(queue)
```
