```python
import copy

original = [[1, 2], [3, 4], [5, 6]]

shallow = copy.copy(original)
deep = copy.deepcopy(original)

original[0][0] = 99

print(shallow[0][0])  # 99 — shallow copy shares the inner list
print(deep[0][0])     # 1  — deep copy is fully independent
```
