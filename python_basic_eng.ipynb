```python
def revers_fun(str_name):
  rev_str =''
  index = len(str_name)
  while index > 0:
    rev_str += str_name[index-1]
    index -= 1
  return rev_str
```

```python
revers_fun(str_name)
```

```python
def unique_list(l):
  x = []
  for a in l:
    if a not in x:
      x.append(a)
  return x

print(unique_list([1,2,3,3,3,3,4,5]))
```

```python
def perfact_num(l):
  x=0
  for i in range(1,l):
    if l%i==0:
      x+=i
  return x==l
print(perfact_num(6))
```
